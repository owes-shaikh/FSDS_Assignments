
### 1. What is the name of the feature responsible for generating Regex objects?

#### Answer:
In Python, the `re.compile()` function helps create what we call "Regex objects." These objects are like tools that let us work with regular expressions, a powerful way to search for and manipulate text.

### 2. Why do raw strings often appear in Regex objects?

#### Answer:
When creating regular expressions, we use raw strings (ones with an 'r' prefix) to make sure Python doesn't misinterpret certain characters. This helps us express patterns precisely in the way we intend.

### 3. What is the return value of the search() method?

#### Answer:
The `search()` method either gives us a "Match" if it finds what we're looking for in the text, or it returns `None` if there's no match.

### 4. From a Match item, how do you get the actual strings that match the pattern?

#### Answer:
To get the actual strings that match our pattern, we use the `group()` method on a Match object. It helps us extract specific parts of the matched text.

### 5. In the regex created from the `r'(\d\d\d)-(\d\d\d-\d\d\d\d)'`, what does group zero cover? Group 2? Group 1?

#### Answer:
In this regex, Group 0 covers the entire matched string, Group 1 corresponds to the first set of parentheses, and Group 2 corresponds to the second set of parentheses.

### 6. In standard expression syntax, parentheses and intervals have distinct meanings. How can you tell a regex that you want it to fit real parentheses and periods?

#### Answer:
To match actual parentheses and periods, we use a backslash (`\`) before them in our regular expression. This tells the regex to treat them as literal characters.

### 7. The `findall()` method returns a string list or a list of string tuples. What causes it to return one of the two options?

#### Answer:
If our regex has groups (defined by parentheses), `findall()` returns a list of tuples where each tuple represents a match and each element in the tuple corresponds to a group. If there are no groups, it returns a list of strings.

### 8. In standard expressions, what does the `|` character mean?

#### Answer:
The `|` character in regular expressions acts like an OR operator. It lets us match either the pattern on its left or the one on its right.

### 9. In regular expressions, what does the character `^` stand for?

#### Answer:
The `^` character in regex signifies the start of a line. It indicates that the pattern following it should match only at the beginning of a line.

### 10. In regular expressions, what is the difference between the `+` and `*` characters?

#### Answer:
- The `+` character means "one or more" of the preceding character or group.
- The `*` character means "zero or more" of the preceding character or group.



### 11. What is the difference between `{4}` and `{4,5}` in regular expression?

#### Answer:
- `{4}` specifies exactly four occurrences of the preceding character or group.
- `{4,5}` specifies a range, allowing for four to five occurrences of the preceding character or group.

### 12. What do you mean by the `\d`, `\w`, and `\s` shorthand character classes signify in regular expressions?

#### Answer:
In regular expressions:
- `\d` matches any digit (0-9).
- `\w` matches any word character (letters, digits, or underscore).
- `\s` matches any whitespace character (spaces, tabs, newlines).

### 13. What do `\D`, `\W`, and `\S` shorthand character classes signify in regular expressions?

#### Answer:
- `\D` matches any non-digit character.
- `\W` matches any non-word character.
- `\S` matches any non-whitespace character.

### 14. What is the difference between `.*?` and `.*`?

#### Answer:
- `.*?` is a non-greedy match that tries to match as little as possible.
- `.*` is a greedy match that tries to match as much as possible.

### 15. What is the syntax for matching both numbers and lowercase letters with a character class?

#### Answer:
To match both numbers and lowercase letters, you can use `[0-9a-z]` or, if you want it to be case-insensitive, `(?i)[0-9a-z]`.

### 16. What is the procedure for making a normal expression in regex case insensitive?

#### Answer:
To make a regex case-insensitive, you can either use the `re.IGNORECASE` flag or insert `(?i)` at the beginning of the regex.

### 17. What does the `.` character normally match? What does it match if `re.DOTALL` is passed as the 2nd argument in `re.compile()`?

#### Answer:
- The `.` character normally matches any character except a newline.
- If `re.DOTALL` is passed, the `.` character matches any character, including a newline.

### 18. If `numReg = re.compile(r'\d+')`, what will `numRegex.sub('X', '11 drummers, 10 pipers, five rings, 4 hen')` return?

#### Answer:
The `sub()` method will replace all occurrences of digits with 'X', resulting in `'X drummers, X pipers, five rings, X hen'`.

### 19. What does passing `re.VERBOSE` as the 2nd argument to `re.compile()` allow you to do?

#### Answer:
Passing `re.VERBOSE` allows you to use whitespace and comments within the regex pattern, making it more readable and organized.

### 20. How would you write a regex that matches a number with a comma for every three digits?

#### Answer:
```python
r'^\d{1,3}(,\d{3})*$'
```
This regex matches numbers with commas for every three digits.

### 21. How would you write a regex that matches the full name of someone whose last name is Watanabe?

#### Answer:
```python
r'^[A-Z][a-z]+\sWatanabe$'
```
This regex matches a full name with the last name 'Watanabe', assuming the first name starts with a capital letter.

### 22. How would you write a regex that matches a sentence with specific words and ends with a period?

#### Answer:
```python
r'^(Alice|Bob|Carol)\s+(eats|pets|throws)\s+(apples|cats|baseballs)\.$'
```
This regex matches sentences with specific word patterns and ends with a period, being case-insensitive.
